```python
import nibabel as nib
import numpy as np
import imageio
import os
import glob
```

```python
results_T1_path = '/content/drive/MyDrive/GP/Clean Dataset/T1 Cropped 3D-T/T1_imgs_middle_only'
results_T2_path = '/content/drive/MyDrive/GP/Clean Dataset/T2 Cropped 3D-T/T2_imgs_middle_only'
results_Flair_path = '/content/drive/MyDrive/GP/Clean Dataset/Flair Cropped 3D-T/Flair_imgs_middle_only'
results_T1c_path = '/content/drive/MyDrive/GP/Clean Dataset/T1c Cropped 3D-T/T1c_imgs_middle_only'

if not os.path.exists(results_T1_path):
    os.makedirs(results_T1_path)

if not os.path.exists(results_T2_path):
    os.makedirs(results_T2_path)

if not os.path.exists(results_Flair_path):
    os.makedirs(results_Flair_path)

if not os.path.exists(results_T1c_path):
    os.makedirs(results_T1c_path)
```

```python
root_path = '/content/drive/MyDrive/GP/BraTS zero artifact'

data_list = sorted(glob.glob(root_path + '/*'))        #list of paths of the inside files
```

```python
padding_flag = 'n'
```

```python
print(sorted(glob.glob(data_list[0] + '/*')))
print(len(data_list))
```

```python
for i in range(len(data_list)):
    modules_list = sorted(glob.glob(data_list[i] + '/*'))          #modules list of a single volume

    flair_vol_path = modules_list[0]                      #path of flair volume is always the first ('sorted')
    t1_vol_path = modules_list[2]                         #path of T1 volume is always the third ('sorted')
    t2_vol_path = modules_list[4]                         #path of T2 volume is always the fifth ('sorted')
    t1c_vol_path = modules_list[3]                         #path of T2 volume is always the forth ('sorted')

    flair_nifti = nib.load(flair_vol_path).get_fdata()                    #Flair volume
    T1_nifti = nib.load(t1_vol_path).get_fdata()                          #T1 volume
    T2_nifti = nib.load(t2_vol_path).get_fdata()                          #T2 volume
    T1c_nifti = nib.load(t1c_vol_path).get_fdata()                          #T2 volume

    for j in range(T1c_nifti.shape[2]):
        flair_nifti_slice = flair_nifti[:, :, j]
        t1_nifti_slice = T1_nifti[:, :, j]
        t2_nifti_slice = T2_nifti[:, :, j]
        t1c_nifti_slice = T1c_nifti[:, :, j]

        if (padding_flag == 'y'):
            flair_nifti_slice = np.pad(flair_nifti_slice, 8)
            t1_nifti_slice = np.pad(t1_nifti_slice, 8)
            t2_nifti_slice = np.pad(t2_nifti_slice, 8)
            t1c_nifti_slice = np.pad(t1c_nifti_slice, 8)

        flair_rslt_path = results_Flair_path + '/' + flair_vol_path.split('/')[-1] + '_' + str(j) + '.png'
        t1_rslt_path = results_T1_path + '/' + t1_vol_path.split('/')[-1] + '_' + str(j) + '.png'
        t2_rslt_path = results_T2_path + '/' + t2_vol_path.split('/')[-1] + '_' + str(j) + '.png'
        t1c_rslt_path = results_T1c_path + '/' + t1c_vol_path.split('/')[-1] + '_' + str(j) + '.png'

        imageio.imwrite(flair_rslt_path, flair_nifti_slice)
        imageio.imwrite(t1_rslt_path, t1_nifti_slice)
        imageio.imwrite(t2_rslt_path, t2_nifti_slice)
        imageio.imwrite(t1c_rslt_path, t1c_nifti_slice)
```
